```python
data[(data['tenure']<tenure_n_th) | (data['tenure']>tenure_p_th)].shape[0]
```

**Observations:**

- There are **175** and **2 observations missing** in **gender** and **tenure** features respectively in the dataset.

- Since the **missing values** are less than **0.18%** of the data, it can be dropped.

- There are **no duplicate values** found in the data.

# **8. Exploratory Data Analysis**
---

- Here we will ask and solve some questions pertaining to our defined problem and on the basis of observations made draw further conclusions

**Question** 
- Which age-group is using facebook the most?

```python
#To find out the frequency of most using age-group we will plot a histogram on age feature

figure = plt.figure(figsize=[15, 7])

b = 10
bins = range(0,116, int(115/b))
data['age'].plot.hist(bins=bins, color='brown')

plt.xticks(ticks=bins, size=10, rotation=90)
plt.yticks(ticks=np.arange(0, 35001, 5000), size=7)
plt.xlabel(xlabel='age', size=14)
plt.ylabel(ylabel='Frequency', size=14)
plt.title(label='Age-wise Distribution of Facebook users', size=16)
plt.show()
```

**Observation**
- Facebook is being used by people ranging from 13 to 113 however the most actively using population belongs to age group 11-33 and there is a sharp decline noticed in usage post 33 years of age followed by declining trend and a little rise between 99-110

**Question**
- What is the percentage of most active age group out of total population?

```python
100*data[(data['age'] >= 11) & (data['age'] <= 33)].shape[0]/data['age'].shape[0]
```

**Observation**
- About 59% of total active population falls in the age-group of 11-33 years

**Question**
- Which gender among the total population is using facebook more?

```python
#To find out the most common gender using facebook we will perform value counts on gender feature and plot a donut chart

figure = plt.figure(figsize=[10, 10])

data['gender'].value_counts().plot(kind='pie',autopct='%3.1f%%',explode=np.ones(2)/10,wedgeprops=dict(width=0.15),fontsize=12,shadow=True,cmap='inferno',label='')

plt.ylabel(ylabel='Gender', size=14)
plt.title(label='Donut Plot showing the usage of facebook by each Gender', size=18)
plt.show()
```

**Observation**
- It shows that out of the total population 59.3% males are using facebook while 40.7% females are using facebook

**Question**
- What is the percentage of male/female in most active age-group using facebook?

```python
figure = plt.figure(figsize=[10, 10])

data[(data['age'] >= 11) & (data['age'] <= 33)]['gender'].value_counts().plot(kind='pie',autopct='%3.1f%%',explode=np.ones(2)/10,wedgeprops=dict(width=0.15),fontsize=12,shadow=True,cmap='inferno',label='')

plt.ylabel(ylabel='Age 11-33', size=14)
plt.title(label='Donut Plot showing Genderwise distribution of most active population', size=18)
plt.show()
```

**Observation**
- Among the most active population, 63.5% are males while 36.5% are females

**Question**
- What is the distribution of dob_month across the population?

```python
data['dob_month'].value_counts().plot.bar()

plt.xlabel('Dob_Month',fontsize=14)
plt.ylabel('Frequency',fontsize=14)
plt.title("Frequency of population by Dob_Month", fontsize=16)
plt.grid(b=True)
plt.show()
```

**Observation**

**Question**
- Which gender is giving/recieving more likes?

```python
#To find out the same we will group by gender feature by sum of the likes given

figure = plt.figure(figsize=[10, 10])
data.groupby(by='gender')['likes'].sum().plot(kind='pie',autopct='%3.1f%%',explode=np.ones(2)/10,wedgeprops=dict(width=0.15),fontsize=12,shadow=True,cmap='inferno',label='')

plt.ylabel(ylabel='Likes Given', size=14)
plt.title(label='Donut Plot showing the likes given by each Gender', size=18)
plt.show()
```

```python
#Similarly perform the same function with likes received feature
figure = plt.figure(figsize=[10, 10])

data.groupby(by='gender')['likes_received'].sum().plot(kind='pie',autopct='%3.1f%%',explode=np.ones(2)/10,wedgeprops=dict(width=0.15),fontsize=12,shadow=True,cmap='inferno',label='')

plt.ylabel(ylabel='Likes Received', size=14)
plt.title(label='Donut Plot showing the likes received by each Gender', size=18)
plt.show()
```

**Observation**
- It shows that 67.9% females are giving likes whereas only 32.1% males are giving likes
- As per the data the likes received by females is 71.8% whereas likes received by males is only 28.2%

**Question**
- What is the genderwise distribution of mobile likes and web likes?

```python
#To find out we will perform groupby on gender and on the sum of mobile and web likes given and store the values in different variables
#We will form a dataframe out of it and plot a bar out of it

figure = plt.figure(figsize=[25, 17])

Mob_likes = data.groupby('gender')['mobile_likes'].sum()
Web_likes = data.groupby('gender')['www_likes'].sum()


like_comp = pd.DataFrame({'Mobile_likes' : Mob_likes, 'Web_likes': Web_likes})

like_comp.plot.bar()
plt.xticks(size=12, rotation=90)
plt.yticks(ticks=np.arange(0, 7500000, 500000), size=14)
plt.xlabel(xlabel='gender', size=14)
plt.ylabel(ylabel='likes', size=14)
plt.title(label='Mobile and Web likes per gender', size=16)
plt.legend(labels=('mobile_likes', 'www_likes'))
plt.grid(b=True)
plt.show()
```

**Observation**
- The graph shows that mobile likes were higher as compared to web likes for both the genders
- It shows that mobile app was used more for liking the posts as compared to web app
- As compared to males, females use web app more likely.

**Question**
- What is the genderwise distribution of likes recieved through mobile and web?

```python
#We will use same function as above here also to find out

Mob_likes_rec = data.groupby('gender')['mobile_likes_received'].sum()
Web_likes_rec = data.groupby('gender')['www_likes_received'].sum()


like_rec_comp = pd.DataFrame({'Mobile_likes_rec' : Mob_likes_rec, 'Web_likes_rec': Web_likes_rec})

like_rec_comp.plot.bar()
plt.xticks(size=12, rotation=90)
plt.yticks(ticks=np.arange(0, 7500000, 500000), size=14)
plt.xlabel(xlabel='gender', size=14)
plt.ylabel(ylabel='likes received', size=14)
plt.title(label='Mobile and Web likes received per gender', size=16)
plt.legend(labels=('mobile_likes_received', 'www_likes_received'))
plt.grid(b=True)
plt.show()
```

**Observation**
- It shows that the number of likes received on mobile were higher than on the likes recieved on web application following the same trend of more likes received by females than by males.

**Question**
- What is the genderwise comparison of Friend Count and Friendships Initiated?

```python
#To find out we will perform groupby on gender and on the sum of Friend count and Friendships initiated and store the values in different variables
#We will form a dataframe out of it and plot a bar out of it

Frnd_ini = data.groupby('gender')['friendships_initiated'].sum()
Frnd_count = data.groupby('gender')['friend_count'].sum()

frnd_comp = pd.DataFrame({'Friendship_initiated' : Frnd_ini, 'Friend_count': Frnd_count})
frnd_comp.plot.bar()
plt.xticks(size=12, rotation=90)
plt.yticks(ticks=np.arange(0, 10000001, 500000), size=14)
plt.xlabel(xlabel='gender', size=14)
plt.ylabel(ylabel='Friendship initiated vs Friend count', size=14)
plt.title(label='Friendship initiated vs Friend count per gender', size=16)
plt.grid(b=True)
plt.show()
```

**Observation**
- More friendships were initiated by males as compared to females however females have higher friend count than males

**Question**
- Which gender has spent longer time on facebook account?

```python
#To find out we will group by gender on mean of tenure feature and plot a bar

data.groupby(by='gender')['tenure'].mean().plot.bar()
plt.xlabel(xlabel='gender', size=14)
plt.ylabel(ylabel='Average Tenure', size=14)
plt.title(label='Average Tenure vs Gender', size=16)
plt.grid(b=True)
plt.show()
```

**Observation**
- It shows that average tenure of females is higher than males

**Question**
- Is there any relation between friend count and frienship initiated?

**Question**
- Is there any relation between friend count and likes given by individuals on various platforms?

```python
#To find out the relation we will plot scatter chart between friend count and likes and doing the same with mobile and web likes respectively
#We will start with scatter plot of friend count and likes

data.plot.scatter(x='friend_count', y='likes', color='purple', figsize=[15, 7])
plt.xlabel(xlabel='friend count', size=14)

plt.ylabel(ylabel='likes', size=14)
plt.title(label='likes vs friend count', size=16)

plt.show()
```

```python
#Now will plot scatter chart between friend count and mobile likes

data.plot.scatter(x='friend_count', y='mobile_likes', color='purple', figsize=[15, 7])
plt.xlabel(xlabel='friend count', size=14)
plt.ylabel(ylabel='mobile likes', size=14)

plt.title(label='mobile likes vs friend count', size=16)

plt.show()
```

```python
#Repeating the same between friend count and web likes

data.plot.scatter(x='friend_count', y='www_likes', color='purple', figsize=[15, 7])

plt.xlabel(xlabel='friend count', size=14)

plt.ylabel(ylabel='web likes', size=14)

plt.title(label='web likes vs friend count', size=16)

plt.show()
```

**Question**
- What is the relation between tenure and friend count if any?

```python
#To find out the relation we will draw a hexbin plot

data.plot.hexbin(x='friend_count', y='tenure', color='#4CB391', figsize=[15, 7], gridsize=25, sharex=False)
plt.yticks(ticks=np.arange(0, 3001, 100), size=12)
plt.xlabel(xlabel='Friend Count', size=14)
plt.ylabel(ylabel='Tenure', size=14)

plt.title(label='Tenure vs Friend count', size=16)
plt.show()
```

**Observation**
- We can observe that the maximum density lies between 300-400.
- People with lesser tenure have more friend count

**Question**
- What is the genderwise comparison of tenure vs friend count?

```python
#We will group by gender on the sum of tenure and friend count and plot a bar graph

tenure = data.groupby('gender')['tenure'].sum()
Frnd_count = data.groupby('gender')['friend_count'].sum()
frnd_comp = pd.DataFrame({'Tenure' : tenure, 'Friend_count': Frnd_count})
frnd_comp.plot.bar()
plt.ylabel(ylabel='tenure vs friend count', size=14)
plt.title(label='Tenure vs Friend count per gender', size=16)
plt.legend(labels=('Tenure', 'Friend count'))
plt.grid(b=True)
plt.show()
```

**Observation**
- It shows that males have higher tenure than females but almost equal to little lesser friend count than females
